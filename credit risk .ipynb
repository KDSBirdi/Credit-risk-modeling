```python
#removing nulls
df1 = df1.loc[df1['Age_Oldest_TL'] != -99999]
```

```python
columns_removed = []

for i in df2.columns:
    if df2.loc[df2[i] == -99999].shape[0] > 10000:
        columns_removed.append(i)
        
df2 = df2.drop(columns_removed, axis=1)
df2
```

```python
for i in df2.columns:
    df2=df2.loc[df2[i]!=-99999]
```

```python
df = pd.merge(df1, df2, how='inner', left_on = ['PROSPECTID'], right_on=['PROSPECTID'])
df
```

```python
for i in df.columns:
    if df[i].dtype == 'object':
        print(i)
```

```python
# !pip install statsmodels
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, precision_recall_fscore_support
```

```python
# chi square test
# FEATURE SELECTION

for i in ['MARITALSTATUS', 'EDUCATION','GENDER','last_prod_enq2','first_prod_enq2','Approved_Flag']:
    chi2,pval, _, _ = chi2_contingency(pd.crosstab(df[i],df['Approved_Flag']))
    print(i, '--', pval)
```

```python
numeric_col = []
for i in df.columns:
    if df[i].dtype != 'object' and i not in ['PROSPECTID','Approved_Flag']:
        numeric_col.append(i)
```

```python
# calculation VIF for multicollinearity 
# VIF = 1 / 1-R2 , Variance Inflation Factor
# VIF = 1 no multicollinearity
# 1-5 Low
# 5-10 Moderate
# > 10 High
```

```python
vif_data = df[numeric_col]
total_col = vif_data.shape[1]
col_to_kept=[]
col_idx=0

for i in range (0, total_col):
    vif_value = variance_inflation_factor(vif_data, col_idx)
    print(col_idx, '--', vif_value)
    
    if vif_value <= 6:
        col_to_kept.append(numeric_col[i])
        col_idx+=1
    else:
        vif_data = vif_data.drop([numeric_col[i]], axis=1)
```

```python
# checking anova with target col

from scipy.stats import f_oneway

col_kept_num = []

for i in col_to_kept:
    a=list(df[i])
    b=list(df['Approved_Flag'])
    
    group_P1 = [value for value, group in zip(a,b) if group == 'P1']
    group_P2 = [value for value, group in zip(a,b)if group == 'P2']    
    group_P3 = [value for value, group in zip(a,b) if group == 'P3']    
    group_P4 = [value for value, group in zip(a,b) if group == 'P4']    
    
    f_stats, p_val = f_oneway(group_P1, group_P2, group_P3, group_P4)
    
    
    if p_val <= 0.05:
        col_kept_num.append(i)
```

```python
features = col_kept_num + ['MARITALSTATUS', 'EDUCATION', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']
df = df[features + ['Approved_Flag']]
```

```python
df.loc[df['EDUCATION'] == 'SSC', ['EDUCATION']] = 1
df.loc[df['EDUCATION'] == '12TH', ['EDUCATION']] = 2
df.loc[df['EDUCATION'] == 'GRADUATE', ['EDUCATION']] = 3
df.loc[df['EDUCATION'] == 'UNDER GRADUATE', ['EDUCATION']] = 3
df.loc[df['EDUCATION'] == 'POST-GRADUATE', ['EDUCATION']] = 4
df.loc[df['EDUCATION'] == 'OTHERS', ['EDUCATION']] = 1
df.loc[df['EDUCATION'] == 'PROFESSIONAL', ['EDUCATION']] = 3
```

```python
df['EDUCATION'].value_counts()
df['EDUCATION'] = df['EDUCATION'].astype(int)
```

```python
df_encoded = pd.get_dummies(df, columns=['MARITALSTATUS', 'GENDER','last_prod_enq2','first_prod_enq2'])
df_encoded
```

```python
df_encoded.info()
```

```python
y = df_encoded['Approved_Flag']
x = df_encoded.drop(['Approved_Flag'], axis=1)
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=200, random_state=42)
```

```python
rf_classifier.fit(x_train,y_train)
y_pred = rf_classifier.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print ()
print(f'Accuracy: {accuracy}')
print ()
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred)


for i, v in enumerate(['p1', 'p2', 'p3', 'p4']):
    print(f"Class {v}:")
    print(f"Precision: {precision[i]}")
    print(f"Recall: {recall[i]}")
    print(f"F1 Score: {f1_score[i]}")
    print()
```
